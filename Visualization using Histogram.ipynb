```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Iris.csv")
print(df)
```

```python
plt.hist(df["SepalLengthCm"],bins=10,color='b',edgecolor='black',alpha=0.7)
plt.xlabel("SepalLengthCm")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
plt.hist(df["SepalWidthCm"],bins=10,color='b',edgecolor='k',alpha=0.7)
plt.xlabel("SepalWidthCm")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Width")
plt.show()
```

```python
plt.hist(df["PetalLengthCm"],bins=10,color='b',edgecolor='k',alpha=0.7)
plt.xlabel("PetalLengthCm")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Length")
plt.show()
```

```python
plt.hist(df["PetalWidthCm"],bins=10,color='b',edgecolor='k',alpha=0.7)
plt.xlabel("PetalWidthCm")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Width")
plt.show()
```

```python
plt.bar(df["Species"],df["Species"].count(),color='b',edgecolor='k')
plt.xlabel("Species")
plt.ylabel("Frequency")
plt.title("Distribution of Species")
plt.show()
```
